## 9 - The most fun part: recognizing the colors on the "faces" of a rubik's cube

```python
# I DIDN'T WRITE THE ENTIRE CODE. IT WAS A CODE MADE IN CLASS.

AREA = 1000

cores = {
    'vermelha':160,
    'verde':60,
    'laranja':10,
    'branco':125,
    'azul':120,
    'amarelo':30
}


def obter_imagem(nome_arquivo):
    return cv2.imread(nome_arquivo)

def converter_imagem_RGB_para_HSV(imagem):
    return cv2.cvtColor(imagem, cv2.COLOR_BGR2HSV)

def exibir_imagem(imagem,titulo):
    cv2.imshow(titulo, imagem)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def retorna_range_cores(valor_cor):
    faixa_inferior = 100

    if valor_cor == cores['branco']:
        faixa_inferior = 8
    tom_claro = np.array([valor_cor - 20, faixa_inferior, 100])
    tom_escuro = np.array([valor_cor + 20, 255, 255])

    return tom_claro, tom_escuro

def obter_imagem_segmentada_por_cor(imagem,valor_cor):
    tom_claro, tom_escuro = retorna_range_cores(valor_cor)
    return cv2.inRange(imagem, tom_claro, tom_escuro)
    

def obter_imagem_binarizada(imagem):
    metodo = cv2.THRESH_BINARY
    ret, imagem_binarizada = cv2.threshold(imagem, 0, 255, metodo)

    #elementos estruturais
    elemento_estruturante_fechamento = cv2.getStructuringElement(cv2.MORPH_RECT,(25,25))
    elemento_estruturante_erosao = cv2.getStructuringElement(cv2.MORPH_RECT,(25,25))
    elemento_estruturante_dilatacao = cv2.getStructuringElement(cv2.MORPH_RECT,(4,4))

    #transformaes morfologicas
    imagem_binarizada = cv2.morphologyEx(imagem_binarizada, cv2.MORPH_CLOSE,elemento_estruturante_fechamento)
    imagem_binarizada = cv2.erode(imagem_binarizada, elemento_estruturante_erosao, iterations = 2)
    imagem_binarizada = cv2.dilate(imagem_binarizada, elemento_estruturante_dilatacao, iterations = 2)
    
    return imagem_binarizada


def obter_contorno_imagem(imagem_binarizada,imagem_original):
    #Obtendo os contornos dos objetos na imagem 
    modo = cv2.RETR_TREE 
    metodo = cv2.CHAIN_APPROX_SIMPLE 
    contornos, hieraquia = cv2.findContours(imagem_binarizada, modo, metodo)

    objeto = None

    resultado = list()

    #Obtendo os contornos do primeiro objeto segmentado 
    if(len(contornos) != 0):
        
        for i in range(0,len(contornos)):
            objeto = contornos[i]
            

            if(cv2.contourArea(objeto) >= AREA):

                # Obtendo os vrtices do retngulo 
                x, y, w, h = cv2.boundingRect(objeto)

                resultado.append([x,y])

                # Desenhando o retngulo na imagem imagemRGB 
                cv2.rectangle(imagem_original, (x,y), (x+w, y+h), (0, 255, 0), 2)

                exibir_imagem(imagem_original, 'Imagem Original')
            
    
    return resultado

def retorna_espaco_cor_HSV():
    brancoRGB = np.uint8([[[255,255,255]]])
    brancoHSV = cv2.cvtColor(brancoRGB, cv2.COLOR_BGR2HSV)
    
def obter_matriz_face(imagem):
    
    imagemRGB = obter_imagem(imagem)
    
    imagemHSV = converter_imagem_RGB_para_HSV(imagemRGB)

    colunas = list()
    linhas = list()
    lista_blocos = list()
    bloco = dict()

    for chave,valor in cores.items():
        imagem_segmentada = obter_imagem_segmentada_por_cor(imagemHSV,valor)
        imagem_binarizada = obter_imagem_binarizada(imagem_segmentada)
        resultadoXY = obter_contorno_imagem(imagem_binarizada,imagemRGB)
        for x,y in resultadoXY:
            colunas.append(x)
            linhas.append(y)
            bloco[chave] = [x,y]
            lista_blocos.append(bloco)
            bloco = dict()

    colunas = sorted(colunas)
    linhas = sorted(linhas)
    
    matriz = [['',''],['','']]

    for bloco in lista_blocos:
        

        for chave,valor in bloco.items():
            c = 0
            l = 0 

            for coluna in colunas:
                if(coluna == valor[0]):
                    if(colunas.index(coluna) > 1):
                        c = 1
                for linha in linhas:
                    if(linha == valor[1]):
                        if(linhas.index(linha) > 1):
                            l = 1
        
            matriz[l][c] = chave
            
            legenda = str(chave)
            cv2.putText(imagemRGB,legenda,(valor[0] + 40,valor[1] + 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 2, cv2.LINE_AA)
        
        
    exibir_imagem(imagemRGB,"Imagem Final")

    return matriz

def detectar_cores():

    imagens = ['/content/drive/MyDrive/Bases_para_Portfolio_e_Testes_py/imagens/cubo_magico.jpeg']
    faces = ['Frontal']
    
    cubo = dict()

    try:
        for i in range(0,len(imagens)):
            cubo[faces[i]] = obter_matriz_face(imagens[i])
        #print(cubo)
    except Exception as e:
        print("Ops aconteceu algum problema!\nO Erro foi: {}".format(e))
```
